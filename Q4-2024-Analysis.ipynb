```python
# Clean LEADS dataframe
leads_df = leads_df.drop(columns=['Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15'])
leads_df['DATE ADDED'] = pd.to_datetime(leads_df['DATE ADDED'], format='%d/%m/%Y', errors='coerce')
leads_df['DOB'] = pd.to_datetime(leads_df['DOB'], errors='coerce')

# Clean SALES dataframe
sales_df = sales_df.drop(columns=['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19'])
sales_df['DATE'] = pd.to_datetime(sales_df['DATE'], format='%d/%m/%Y', errors='coerce')

# Check for any remaining NaN values or issues
print(leads_df.info())
print(sales_df.info())
```

```python
# Convert 'DATE ADDED' to datetime format (you may need to adjust the format based on the actual format)
leads_df['DATE ADDED'] = pd.to_datetime(leads_df['DATE ADDED'], format='%d/%m/%Y', errors='coerce')

# Convert 'DOB' to datetime format
leads_df['DOB'] = pd.to_datetime(leads_df['DOB'], errors='coerce')

# Convert 'DATE' to datetime format
sales_df['DATE'] = pd.to_datetime(sales_df['DATE'], format='%d/%m/%Y', errors='coerce')

# Check for any remaining NaN values or issues
print(leads_df.info())
print(sales_df.info())
```

```python
# Clean LEADS dataframe
leads_df_clean = leads_df.dropna(subset=['DOB', 'DATE ADDED'])
leads_df_clean['DOB'] = pd.to_datetime(leads_df_clean['DOB'], errors='coerce')
leads_df_clean['DATE ADDED'] = pd.to_datetime(leads_df_clean['DATE ADDED'], errors='coerce')

# Clean SALES dataframe
sales_df['DOB'] = pd.to_datetime(sales_df['DOB'], errors='coerce')
sales_df = sales_df.dropna(subset=['NUMBER'])

# Drop duplicates (if needed)
leads_df_clean = leads_df_clean.drop_duplicates()
sales_df_clean = sales_df.drop_duplicates()

# Check the cleaned info
print(leads_df_clean.info())
print(sales_df_clean.info())



```

```python
leads_df_clean.loc[:, 'DOB'] = pd.to_datetime(leads_df_clean['DOB'], errors='coerce')
leads_df_clean.loc[:, 'DATE ADDED'] = pd.to_datetime(leads_df_clean['DATE ADDED'], errors='coerce')
```

```python
# Demographics Analysis
# Calculate Age
current_date = pd.to_datetime('today')

# For Leads DataFrame
leads_df_clean['AGE'] = (current_date - leads_df_clean['DOB']).dt.days // 365

# For Sales DataFrame
sales_df['AGE'] = (current_date - sales_df['DOB']).dt.days // 365

# Age groups: 18-30, 31-40, 41-50, 51-60, 61+
bins = [0, 30, 40, 50, 60, 100]
labels = ['18-30', '31-40', '41-50', '51-60', '61+']
leads_df_clean['AGE GROUP'] = pd.cut(leads_df_clean['AGE'], bins=bins, labels=labels, right=False)

# Create AGE GROUP for Sales DataFrame
sales_df['AGE GROUP'] = pd.cut(sales_df['AGE'], bins=bins, labels=labels, right=False)

# Age group distribution for leads and sales
leads_age_distribution = leads_df_clean['AGE GROUP'].value_counts().sort_index()
sales_age_distribution = sales_df['AGE GROUP'].value_counts().sort_index()

# Display the age group distribution
print("Leads Age Group Distribution:\n", leads_age_distribution)
print("Sales Age Group Distribution:\n", sales_age_distribution)
```

```python
# Insurance Type Distribution
leads_insurance_distribution = leads_df_clean['INSURENCE COMPANY'].value_counts()
sales_insurance_distribution = sales_df['INSURANCE COMPANY'].value_counts()

# Display the insurance distribution
print("Leads Insurance Type Distribution:\n", leads_insurance_distribution)
print("Sales Insurance Type Distribution:\n", sales_insurance_distribution)
```

```python
# Tier Distribution
leads_tier_distribution = leads_df_clean['TIER'].value_counts()
sales_tier_distribution = sales_df['TIER'].value_counts()

# Display the tier distribution
print("Leads Tier Distribution:\n", leads_tier_distribution)
print("Sales Tier Distribution:\n", sales_tier_distribution)
```

```python
# Check column names for both DataFrames
print("Leads Columns:", leads_df_clean.columns)
print("Sales Columns:", sales_df.columns)

# Clean column names by stripping extra spaces
leads_df_clean.columns = leads_df_clean.columns.str.strip()
sales_df.columns = sales_df.columns.str.strip()

# Convert 'PATIENT RESPONSIBILITY' to numeric (handle errors as NaN)
leads_df_clean['PATIENT RESPONSIBILITY'] = pd.to_numeric(leads_df_clean['PATIENT RESPONSIBILITY'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')
sales_df['PATIENT RESPONSIBILITY'] = pd.to_numeric(sales_df['PATIENT RESPONSIBILITY'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')

```

```python
# Calculate the average responsibility for leads and sales
leads_avg_responsibility = leads_df_clean['PATIENT RESPONSIBILITY'].mean()
sales_avg_responsibility = sales_df['PATIENT RESPONSIBILITY'].mean()

print(f"Leads Average Patient Responsibility: ${leads_avg_responsibility:.2f}")
print(f"Sales Average Patient Responsibility: ${sales_avg_responsibility:.2f}")
```

```python
# Group sales data by INSURANCE COMPANY and count the number of sales
insurance_sales = sales_df.groupby('INSURANCE COMPANY')['STATUS'].count().reset_index()

# Sort the sales by count in descending order to find the top-selling insurance company
insurance_sales_sorted = insurance_sales.sort_values(by='STATUS', ascending=False)

# Display the top-selling insurance companies
top_insurance_sales = insurance_sales_sorted.head(10)  # Display top 10

print(top_insurance_sales)

# Optional: Visualize the data (Top 10 insurance companies with most sales)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.barh(top_insurance_sales['INSURANCE COMPANY'], top_insurance_sales['STATUS'], color='skyblue')
plt.xlabel('Number of Sales')
plt.title('Top 10 Insurance Companies with the Most Sales')
plt.gca().invert_yaxis()  # To show the highest at the top
plt.show()
```

```python
# Count age groups for leads and sales
leads_age_group = leads_df_clean['AGE GROUP'].value_counts().sort_index()
sales_age_group = sales_df['AGE GROUP'].value_counts().sort_index()

# Create a DataFrame for comparison
age_group_comparison = pd.DataFrame({
    'Leads': leads_age_group,
    'Sales': sales_age_group
}).fillna(0)

# Plot the comparison
age_group_comparison.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'salmon'])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Leads vs Sales by Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define the replacements for the 'TIER' column in sales_df
tier_replacements = {
    '0-75 $': 'TIER 1',
    '200-329 $': 'TIER 3',
    '76-199 $': 'TIER 2'
}

# Apply the replacements to the 'TIER' column in sales_df
sales_df['TIER'] = sales_df['TIER'].replace(tier_replacements)

# Verify the changes
print(sales_df['TIER'].value_counts())
```

```python
# Group by TIER and calculate the average patient responsibility
avg_responsibility_by_tier_leads = leads_df_clean.groupby('TIER')['PATIENT RESPONSIBILITY'].mean()
avg_responsibility_by_tier_sales = sales_df.groupby('TIER')['PATIENT RESPONSIBILITY'].mean()

# Create a DataFrame for comparison
responsibility_comparison = pd.DataFrame({
    'Leads': avg_responsibility_by_tier_leads,
    'Sales': avg_responsibility_by_tier_sales
})

# Plot the comparison
responsibility_comparison.plot(kind='bar', figsize=(10, 6), color=['lightgreen', 'orange'])
plt.xlabel('Tier')
plt.ylabel('Average Patient Responsibility ($)')
plt.title('Average Patient Responsibility by Tier')
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot histograms for Leads and Sales Patient Responsibility
plt.figure(figsize=(12, 6))
plt.hist(leads_df_clean['PATIENT RESPONSIBILITY'], bins=30, alpha=0.5, label='Leads', color='skyblue')
plt.hist(sales_df['PATIENT RESPONSIBILITY'], bins=30, alpha=0.5, label='Sales', color='salmon')
plt.legend()
plt.xlabel('Patient Responsibility ($)')
plt.ylabel('Frequency')
plt.title('Patient Responsibility Distribution for Leads vs Sales')
plt.show()
```

```python
# Group sales data by INSURANCE COMPANY and calculate the percentage of sales
insurance_sales_percentage = sales_df['INSURANCE COMPANY'].value_counts(normalize=True).head(10) * 100

# Plot the pie chart
plt.figure(figsize=(8, 8))
insurance_sales_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 10 Insurance Companies by Sales Percentage')
plt.ylabel('')
plt.show()
```

```python
# Group by insurance company and calculate the average patient responsibility
avg_responsibility_by_insurance_leads = leads_df_clean.groupby('INSURENCE COMPANY')['PATIENT RESPONSIBILITY'].mean().sort_values(ascending=False)
avg_responsibility_by_insurance_sales = sales_df.groupby('INSURANCE COMPANY')['PATIENT RESPONSIBILITY'].mean().sort_values(ascending=False)

# Create a DataFrame for comparison
responsibility_by_insurance = pd.DataFrame({
    'Leads': avg_responsibility_by_insurance_leads,
    'Sales': avg_responsibility_by_insurance_sales
}).head(10)  # Only top 10 for clarity

# Plot the comparison
responsibility_by_insurance.plot(kind='barh', figsize=(12, 8), color=['lightblue', 'salmon'])
plt.xlabel('Average Patient Responsibility ($)')
plt.title('Top 10 Insurance Companies by Average Patient Responsibility')
plt.gca().invert_yaxis()  # Highest at the top
plt.show()
```

```python
# Calculate the percentage distribution of sales by tier
sales_tier_distribution = sales_df['TIER'].value_counts(normalize=True) * 100

# Plot the pie chart
plt.figure(figsize=(8, 8))
sales_tier_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Sales Distribution by Tier')
plt.ylabel('')
plt.show()
```
